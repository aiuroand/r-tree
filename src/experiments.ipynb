```python
import matplotlib.pyplot as plt
```

```python
n                     = [100, 1000, 10000, 100000]
insert_brute          = [0.496833, 4.239704, 47.224586, 394.495914]
insert_rtree_qudratic = [4.061355, 63.769009, 847.097004, 10441.859968]
insert_rtree_linear   = [2.719477, 45.638028, 598.402382, 8065.307693 ]

fig = plt.figure(figsize=(4,3))

plt.plot(n, insert_brute, label="Brute Force")
plt.plot(n, insert_rtree_qudratic, label="R-tree + Quadratic split")
plt.plot(n, insert_rtree_linear, label="R-tree + Linear split")
plt.ylabel("Time [ms]")
plt.xlabel("Number of records")
plt.xscale('log')
plt.title("Insert")
plt.legend()
plt.show()
```
